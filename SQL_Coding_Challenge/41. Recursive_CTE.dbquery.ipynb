```sql
-- Recursive CTE

with cte_numbers as
(
  select 1 as num  -- Anchor query

  union all

  select num + 1   -- Recursive query
  from cte_numbers
  where num < 10   -- Filter to stop the recursion

)
select num 
from cte_numbers;

-- Databricks Syntax:
select explode(
  sequence(1, 10, 1)
) as num





------------------------------ TABLE FORMATION ------------------------------
-- 1. Create the tables
drop table if exists sales;

create table sales (
product_id int,
period_start date,
period_end date,
average_daily_sales int
);

insert into sales values(1,'2019-01-25','2019-02-28',100),(2,'2018-12-01','2020-01-01',10),(3,'2019-12-01','2020-01-31',1);

------------------------------------------------------------------------------------------

/*
Q1: User purchase platform.

*/

-- Ans1:
-- Method 1: 
select * from sales;



```
