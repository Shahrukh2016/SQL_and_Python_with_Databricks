```sql
------------------------------ TABLE FORMATION ------------------------------
drop table if exists entries;

create table entries ( 
name varchar(20),
address varchar(20),
email varchar(20),
floor int,
resources varchar(10));

insert into entries 
values ('A','Bangalore','A@gmail.com',1,'CPU'),('A','Bangalore','A1@gmail.com',1,'CPU'),('A','Bangalore','A2@gmail.com',2,'DESKTOP')
,('B','Bangalore','B@gmail.com',2,'DESKTOP'),('B','Bangalore','B1@gmail.com',2,'DESKTOP'),('B','Bangalore','B2@gmail.com',1,'MONITOR');


------------------------------------------------------------------------------------------

/*
Q1:
output: name, total_visits, most_visited_floor, resources_used
*/



-- Ans1:
-- Method 1: 
select * from entries;

with 
total_visits as (
  select name, count(1) as total_visits, string_agg(distinct resources, ',') as resources_used 
  from entries
  group by name
  ),

floor_visit as (
  select name, floor, count(1) as no_of_floor_visit,
  rank() over(partition by name order by count(1) desc) as rn
  from entries
  group by name, floor
  )

select floor_visit.name, total_visits.total_visits, floor_visit.floor as mots_visited_floor, total_visits.resources_used
from floor_visit
join total_visits
on floor_visit.name = total_visits.name
where rn = 1;

```
