```sql
------------------------------ TABLE FORMATION ------------------------------
-- 1. Create the tables
drop table if exists orders;
drop table if exists products;

create table orders
(
order_id int,
customer_id int,
product_id int
);

insert into orders VALUES 
(1, 1, 1),
(1, 1, 2),
(1, 1, 3),
(2, 2, 1),
(2, 2, 2),
(2, 2, 4),
(3, 1, 5);

create table products (
id int,
name varchar(10)
);
insert into products VALUES 
(1, 'A'),
(2, 'B'),
(3, 'C'),
(4, 'D'),
(5, 'E');
------------------------------------------------------------------------------------------

/*
Q1: Find total sales by year.
*/

-- Ans1:
-- Method 1: 
select * from orders;
select * from products;

select o1.order_id, o1.product_id as p1, o2.product_id as p2
from orders as o1
inner join orders as o2
on o1.order_id = o2.order_id and o1.product_id <> o2.product_id;









select o.order_id, string_agg(p.name)
from orders as o
left join products as p
on o.order_id = p.id
group by o.order_id
order by o.order_id;






```
