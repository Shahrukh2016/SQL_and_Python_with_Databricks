```sql
------------------------------ TABLE FORMATION ------------------------------
-- 1. Create the tables
drop table if exists person;
drop table if exists friend;

CREATE TABLE person (
    PersonID INT,
    Name VARCHAR(50),
    Email VARCHAR(100),
    Score INT,
    PRIMARY KEY (PersonID)
);

CREATE TABLE friend (
    PersonID INT,
    FriendID INT,
    PRIMARY KEY (PersonID, FriendID)
);

-- 2. Insert data into person
INSERT INTO person (PersonID, Name, Email, Score) VALUES
(1, 'Alice', 'alice2018@hotmail.com', 88),
(2, 'Bob', 'bob2018@hotmail.com', 11),
(3, 'Davis', 'davis2018@hotmail.com', 27),
(4, 'Tara', 'tara2018@hotmail.com', 45),
(5, 'John', 'john2018@hotmail.com', 63);

-- 3. Insert data into friend
INSERT INTO friend (PersonID, FriendID) VALUES
(1, 2), (1, 3), 
(2, 1), (2, 3), 
(3, 5), 
(4, 2), (4, 3), (4, 5);

------------------------------------------------------------------------------------------

/*
Q1:
Write a query to find PersonID, Name, number of friends, sum of marks of person who have friends with total score greater than 100.
*/

-- Ans1:
-- Method 1: 
select * from person;
select * from friend;

with qualified_persons as (
select friend.personid, count(friend.friendid) as total_friends, sum(person.score) as friends_score 
from friend
left join person
on friend.friendid = person.personid
group by friend.personid
)
select qualified_persons.personid, person.name, qualified_persons.total_friends as no_of_friends, qualified_persons.friends_score as sum_of_marks 
from qualified_persons
join person
on qualified_persons.personid = person.personid
where friends_score > 100;

```
