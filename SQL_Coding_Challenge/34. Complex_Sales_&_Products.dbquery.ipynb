```sql
------------------------------ TABLE FORMATION ------------------------------
/*
Conept:
The Pareto principle states that for many outcomes, roughly 80% of consequences come from 20% of causes. eg:
1-80% of the productivity come from 20% of the employees.
2-80% of your sales come from 20% of your clients.
3-80% of decisions in a meeting are made in 20% of the time
4-80% of your sales comes from 20% of your products or services.

Table Used:
practice_datasets.default.superstore_orders
*/


------------------------------------------------------------------------------------------

/*
Q1:
Product ids that follows Pareto principle.
*/



-- Ans1:
SELECT * FROM practice_datasets.default.superstore_orders limit 10;

with eighty_perc_sales as 
(
select product_id ,sum(sales) as sales, (SELECT sum(sales) * 0.8 FROM practice_datasets.default.superstore_orders) as eighty_perc
from practice_datasets.default.superstore_orders 
group by product_id
order by sales desc
),
cumulative_sales as
(
select *, sum(sales) over (order by sales desc) as cumulative_sales 
from eighty_perc_sales 
)
select * from cumulative_sales
where cumulative_sales <= eighty_perc;

```
