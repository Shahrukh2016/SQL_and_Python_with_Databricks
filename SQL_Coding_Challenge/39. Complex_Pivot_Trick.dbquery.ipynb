```sql
------------------------------ TABLE FORMATION ------------------------------
-- 1. Create the tables
drop table if exists tasks;

create table tasks (
date_value date,
state varchar(10)
);

insert into tasks values 
('2019-01-01','success'),('2019-01-02','success'),
('2019-01-03','success'),('2019-01-04','fail'),
('2019-01-05','fail'),('2019-01-06','success');
------------------------------------------------------------------------------------------

/*
Q1:
output: start_date, end_date, state.
*/

-- Ans1:
-- Method 1: 
select * from tasks;

with cte as (
select *, row_number() over(partition by state order by date_value asc) as group_date,
dateadd(day, -1*row_number() over(partition by state order by date_value asc), date_value) as groups
from tasks
order by date_value
)
select min(date_value) as start_date, max(date_value) as end_date, state
from cte
group by groups, state
order by start_date;













```
